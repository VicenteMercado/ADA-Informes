#1. Problema del ordenamiento.

**Entrada:** Una secuencia de n números $[x_1,x_2,...,x_n]$

**Salida:** Permutación ordenada de la secuencia de entrada: $[x_1',x_2',...,x_n']$, de tal manera que $x_1'\leq x_2' \leq... \leq x_n'$.

![image](https://codekyro.com/wp-content/uploads/2021/05/insertion-sort.jpg)

Desde los inicios de la computación hasta la actualidad, uno de los planteamientos que más atracción ha generado sin dudas ha sido el problema del ordenamiento, el cual pese a ser un concepto bastante básico a plena vista, ha demostrado ser bastante complicado de solucionar de manera eficiente, y hasta el día de hoy se siguen generando nuevos algoritmos de ordenamiento para satisfacer esta problemática. Un ejemplo conocido de estos algoritmos es InsertionSort, el cual ha sido fue mencionado por primera vez y analizado desde 1946 en adelante.




#2. InsertionSort

El código a continuación muestra una implementación del algoritmo **InsertionSort**.

##2.1 Código.

```python
import random
from termcolor import colored

def insertionSort(arr, verbose = False):
    
    if verbose == True: print("Input array:", arr)
    cont = 0 #Contador de retrocesos entre todas las iteraciones.
    N = 0 #Contador de comparaciones.

    for i in range(1, len(arr)):
        
        if verbose == True: print("\nPasada ", i)

        key = arr[i]
        j = i - 1
        
        # Comparar cada posición con su cercana por la izquierda para ver si esta última es mayor que la primera o no.
        while j >= 0 and key < arr[j]:
            aux = arr[j+1]
            arr[j + 1] = arr[j]
            arr[j] = aux
            j = j - 1
            cont+=1

            if verbose == True:
              print(arr)
        
        # Actualizar key hacia la siguiente posición a evaluar.
        arr[j + 1] = key

        if verbose == True:
              print("Al finalizar recorrido: ", arr)
    
    if verbose == True: print("\noutput array:",arr)

    T = cont / i #Media de retrocesos.
    return arr, T, cont



#Ejemplo
arr = [58, 15, 64, 30, 25, 67]
print("Entrada: ",arr)
arr, T,comp = insertionSort(arr)
print("Salida: ",arr)
print("# Comparaciones: ", comp)
print("Media de retrocesos: ", T)
```

##2.2 Descripción del algoritmo.

El algoritmo recibe como entrada una lista (o arreglo) $arr$  con la secuencia de $n$  elementos que se desean ordenar. Luego, los números se ordenan dentro de la misma lista.
 
1. En cada iteración $i$ del arreglo, se compara cada posición `arr[j]` con su predecesor `[j-1]` para ver si respetan el sentido del orden. Si sucede que `arr[j]`<`arr[j-1]`, los elementos intercambian de posiciones.

2. Si el arreglo se halla totalmente ordenado, el algoritmo termina. En caso contrario, se recorre nuevamente acorde al *paso 1*.

3. Finalmente, se retorna el arreglo ordenado y la media de retrocesos realizados en el proceso.

$n$ será el tamaño del problema en cuestión.

Cuando la variable `verbose` es `True` se muestra información para ver lo que pasa paso a paso dentro de la función.

##2.3 Ejemplo.

Considere el siguiente arreglo:

$x = [4, 2, 7, 9, 5]$

Para la primera iteración, se intercambian el 4 por el 2.

$x = [2, 4, 7, 9, 5]$

Para la segunda iteración, no sucede un intercambio de números, al no haber números mayores a 7 por la izquierda. Lo mismo va para la tercera iteración con 9.

Para la cuarta iteración, el 5 intercambia posiciones con el 9, y después con el 7.

$x = [2, 4, 5, 7, 9]$

Tenemos, por último, nuestro arreglo ordenado.

#3. Tiempo de ejecución

# Teorema (Tiempo de ejecución).
*El algoritmo InsertionSort tiene un tiempo de ejecución de  $O(n^2)$  en el peor caso.*

# Prueba del teorema
Si se contempla detenidamente el código, nos daremos cuenta de que el peor caso posible se dará cuando el menor elemento se encuentra en la última posición del arreglo. De esta forma, con cada recorrido se movería una casilla a la izquierda hasta llegar a la primera posición, por lo que es necesario llevar $n-1$ recorridos a cabo para que los elementos estén ordenados.

##Ejemplo

```python
a = [4,3,2,1]
print("Entrada no ordenada:",a)
a,T,comp=insertionSort(a,verbose=True)
print("Salida ordenada:",a)
print("# Comparaciones: ", comp)
print("Media de retrocesos: ", T)
```

Para esta situación, en la primera pasada se realizó 1 comparación, en la segunda 2 y en la tercera 3. Un total de 6 comparaciones.

En el caso general, se deberían realizar  $\sum\limits_{i=0}^{n-1} i=\frac{n(n-1)}{2}$  comparaciones. Por lo que **el tiempo de ejecución del algoritmo en el peor caso** es  $O(n^2)$. 

#4. Correctitud

## Teorema (Correctitud).

*El algoritmo **InsertionSort** genera un arreglo:  [x′1,x′2,...,x′n] , con los mismos elementos del arreglo de entrada ordenados de menor a mayor, es decir,  $x′1≤x′2≤...≤x′n$ .*

## Prueba del Teorema.

Previamente observamos que en cada iteración de InsertionSort se añade un nuevo elemento ordenado al inicio del arreglo.

Por lo que probaremos la siguiente **propiedad invariante de bucle** para el `InsertionSort`:

> Al comienzo de cada iteración del bucle más grande, el subarreglo $a[0,...,i-1]$ contiene **los mismos elementos** que el arreglo original, pero **ordenados de menor a mayor**.

**Inicialización**

Al inicio de la primera iteración (i=1), el subarreglo $[x_0]$ contiene a la posición $i-1=0$ y cumple la propiedad, lo cual es trivialmente cierto.

**Mantención**

A partir de la conclusión anterior, podemos asumir que al principio de cada iteración $i$ se cumple la propiedad, es decir, el subarreglo $a[0,...,i-1]$ contiene los mismos elementos que el arreglo original, pero ordenados de menor a mayor.

##5.1 Número de comparaciones

A continuación se muestra gráficamente una comparación entre:

* cantidad de comparaciones del peor caso calculadas matemáticamente,
* cantidad de comparaciones del mejor caso calculadas matemáticamente ($n-1$ comparaciones si el arreglo ya está ordenado),
* cantidad de comparaciones realizadas experimentalmente para entradas aleatorias.

Se tomarán en cuenta tamaños de problemas con $n \in [10,29]$.

Los arreglos de entrada son generados aleatoriamente haciendo un sampling de un conjunto de números entre 1 y 100.

```python
import matplotlib.pyplot as plt

x=[n for n in range(10,30)] 
y1=[n*(n-1)/2 for n in range(10,30)] # worst case
y2=[n-1 for n in range(10,30)] # best case
y=[]; 

for n in range(10,30):
  a = random.sample(range(1, 100), n)
  a,T,comp = insertionSort(a)
  y.append(comp)

plt.plot(x,y)
plt.plot(x,y1)
plt.plot(x,y2)
plt.legend(["Insertion Sort", "Peor caso", "Mejor caso"])

plt.xlabel('n')
plt.ylabel('number of operations')
plt.show()
```

##Análisis de resultados

En el gráfico, y tal como era de esperarse, podemos ver que el tiempo de ejecución promedio está por debajo del peor caso teórico.
El mejor caso teórico posee una notación asintótica $O(n)$, aunque los casos de arreglos generados al azar tienden más hacia el promedio.
