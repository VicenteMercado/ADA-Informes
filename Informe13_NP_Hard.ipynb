#1. Introducción
Previamente analizamos distintos problemas, cada uno con sus respectivas entradas y salidas, y un algoritmo capaz de resolverlo en algún tiempo estimado, dependiendo de los métodos utilizados. Sin embargo, esto no siempre es así, y la verdad es que no todos los problemas pueden ser resueltos tan **fácilmente**. Estos son los llamados problemas **NP-Hard**, y los más importante sobre ellos será analizado durante este informe.

##1.1 ¿Qué quiere decir que un problema sea NP-Hard?

Que un problema sea **NP-Hard** quiere decir que cualquier problema $NP$ puede ser reducido a este. En otras palabras, es al menos tan difícil como cualquier problema con soluciones fácilmente reconocibles.

Un problema será NP-Hard, asumiendo que $P \neq NP$ es cierto, si este no puede ser resuelto en un tiempo polinomial, ni siquiera por un algoritmo con un tiempo de ejecución $O(n^{100})$ ó incluso $O(n^{1000})$ para un tamaño del problema $n$. 

Si es que esto fuera posible, entonces implicaría que la conjetura $P \neq NP$ sería falsa y generaría un algoritmo en tiempo polinomial para todo problema cuya solución es reconocible en tiempo polinomial, lo cual sería un caso demasiado generoso para ser cierto.

##1.2 ¿Qué quiere decir que $P \neq NP$?
Para poder definir esta famosa conjetura en el mundo de la informática, es necesario primero definir que son $P$ y $NP$ respectivamente. Ambos conceptos nacen desde la teoría de la complejidad computacional, y hacen referencia a problemas que pueden ser resueltos desde cero en un tiempo polinomial y aquellos cuyas soluciones pueden ser verificadas en un tiempo polinomial respectivamente.

Siguiendo estas definiciones, en general podemos formar la siguiente respuesta a nuestra pregunta inicial:
> $P$ es diferente a $NP$ debido a que chequear una presunta solución a un problema puede ser fundamentalmente más sencillo que formar una solución desde cero.

Así, y podemos definir que cada problema $P$ también pertenece a $NP$, es decir, $P ⊆ NP$, y que, por lo tanto, una definición más concisa y "formal" sería la siguiente:

> $P$ es un subconjunto estricto de $NP$.

image.png

Esta conjetura afirma la existencia de problemas con soluciones eficientes que se pueden reconocer (es decir, $NP$), pero que no puede ser resuelto por un algoritmo de tiempo polinomial. Si ésta fuera falsa, si un problema es $NP$ entonces siempre habrá un algoritmo que pueda solucionarlo, sin embargo, esto generaría un conflicto con lo descrito anteriormente acerca de NP-Hard, ya que un algoritmo de tiempo polinomial para un problema de este tipo llevaría directamente a uno para cada problema $NP$, probando que $P = NP$ y refutando a $P \neq NP$. De esta forma, podemos concluir lo siguiente:

> Si $P \neq NP$ es verdad, ningún problema NP-Hard puede ser resuelto por un algoritmo de tiempo polinomial.

##1.3 ¿Cómo se prueba que un problema es NP-Hard usando **reducciones**?
**Reducir** en este contexto, significa transformar una instancia de un problema $A$ a una instancia de problema $B$, esta transformación teniendo un tiempo polinomial.

Si es posible realizar esta transformación, y se sabe que $A$ es NP-Hard, entonces $B$ tiene que ser NP-Hard, por que si no, surge una contradicción. Si transformamos a $A$ en un problema de $B$ y lo resolvemos con tiempo polinomial, entonces $A$ no puede ser NP-Hard. Por lo tanto, $B$ tiene que ser NP-hard.

De ahí surge la manera de realizar la comprobación de NP-Hard. Tomamos un problema $A$, que sabemos con certeza que es NP-Hard, y buscamos la forma de transformarlo al problema $B$, el cual queremos comprobar si es NP-Hard o no.

##2.1 Problema del cliqué

###2.1.1 Descripción del problema

**Entrada:** Un Grafo no dirigido $G = (V,E)$

**Salida:** El subgrafo $G' \in G$, donde los nodos de $G'$ forman el cliqué más grande del nodo.

Un cliqué se define como un conjunto de nodos en un grafo no dirigido, donde los nodos están conectado con cada uno de los demás miembros del cliqué mediante arcos, como se puede ver en el grafo siguiente:

Como se puede ver destacado en el grafo, el subconjunto de vértices $5,2,1$ es el cliqué más grande con un tamaño de $3$ vértices. Cabe mencionar que también existen otros cliqués presentes que serían los nodos adyacentes, los cuales son de tamaño 2 ($6,4; 4,3; etc.$). No obstante, no satisfacen la salida del problema al no ser los cliqués de mayor tamaño.

###2.1.2 ¿Por qué es NP-Hard?

###Teorema (NP-Hardness del problema del cliqué)

> *El problema del cliqué es NP-Hard.*

Para probar nuestro teorema, usaremos el principio de **reducción**, donde el problema presente $B$ sería el resultado de reducir un problema $A$, sabiendo que $A$ es **NP-Hard**. En nuestro caso, el problema $A$ será el **problema del conjunto independiente máximo**, el cual intentaremos reducir al problema $B$, es decir, el **problema del cliqué**.

image.png

###Prueba del teorema

Para comprobar si nuestro problema $A$ se puede, en efecto, reducir al problema $B$, hay que describir formalmente en qué consiste. Similar al problema del cliqué, el problema del conjunto independiente máximo recibe un grafo no dirigido $G = (V,E)$, donde dado éste se le denomina **conjunto independiente** a aquel subconjunto de nodos $S ⊆ V$ donde no hay dos o más nodos que se conecten mediante un arco $e ⊆ E$. Por lo que el problema retornaría el mayor conjunto independiente que se puede obtener desde el grafo $G$, es decir, un valor $k$ que correspondería a la cantidad de nodos que conforman $S$.

image.png

Ambos problemas reciben paramétros iguales, siendo un grafo no dirigido $G=(V,E)$, aunque los problemas $A$ y $B$ intentan resolver incógnitas opuestas (nodos **no adyacentes** y nodos **unidos entre sí**, respectivamente).

Consideremos ahora un grafo $G'$ similar a $G$, con los mismos nodos pero con la diferencia de que cada par de nodos conectados por un arco en $G$ está desconectado en $G'$ y viceversa. Este se llama el grafo complemento de $G$. 

![image](https://static.javatpoint.com/tutorial/dms/images/complement-of-graph-in-discrete-mathematics.png)

Analizando este grafo, podemos deducir que un set independiente en $G$ se convierte en un cliqué en el grafo complemento $G′$, ya que al no estar conectados en el grafo original implica que estén conectados completamente en el complemento. Por lo tanto, resolver el problema del set independiente es equivalente a resolver el problema del clique con el grafo complemento como entrada. Vale decir que el traspaso de un grafo a su complemento tiene tiempo polinomial, ya que a lo más habrán $n*(n-1)/2$ arcos para un grafo de $n$ nodos.

Ya que el problema del set independiente es NP-Hard, y sabiendo que se puede reducir al problema del cliqué en tiempo polinomial, podemos concluir que el problema del cliqué es **NP-Hard**.

Ahora que sabemos que la reducción de nuestro problema $A$ al problema $B$ es efectiva, es posible crear un algoritmo que pueda hallar el conjunto independiente máximo siguiendo la estructura descrita.

1. Nuestro preproceso sería crear un grafo complemento $G'$ a partir del grafo de entrada $G$.

2. Resolvemos el problema del cliqué máximo posible dentro de $G'$.

3. Para el postproceso y como los $k$ vértices en el grafo complementario conforman el cliqué máximo, entonces quiere decir que en el grafo original compondrán el set independiente máximo, por lo que retornamos dichos vértices.

##2.1.3 Algoritmo sencillo para resolver el problema.

Si bien nunca existirá un algoritmo que retorne una solución eficiente para el problema, podemos proponer uno que entregue un resultado aproximado. Como entrada tendría un grafo no dirigido $G = (V,E)$ y consistiría de los siguientes pasos:

1. Definir a un contador $k$ de tamaño inicial 2, y un arreglo vacío para almacenar los cliques de tamaño $k$.

2. Iteramos sobre cada nodo, guardando un cliqué que lo contenga junto a uno de sus nodos adyacentes.

3. Incrementamos a $k$ en uno, y creamos un nuevo arreglo para cliques de este tamaño $k=3$. Iteramos sobre el arreglo de cliques tamaño $2$ anterior, chequeando si para cada clique, agregar cierto nodo adjacente a los nodos del clique todavía resultaría en un clique. Si se cumple, se agrega.

4. Se continua incrementando $k$ y realizando el paso anterior hasta que no se pueda crear un cliqué de tamaño $k$. Se retorna $k-1$, que sería el maximo tamaño de cliqué encontrado.

Para un algoritmo tan sencillo como esto, vemos que su tiempo de ejecución es de $O(2^n)$, ya que por cada nodo habría que revisar todos los nodos adyacentes, y esto se realiza para cada tamaño $k$ de cliqué a buscar.


##2.2 Problema de la cobertura de vértices

###2.2.1 Descripción del problema
**Entrada:** Un grafo no dirigido $G = (V, E)$.

**Salida:** Un valor $k \in \mathbb{R}$ que corresponde al tamaño del subconjunto $S$ que contiene la menor cobertura de vértices.

En la teoría de grafos, se define a una **cobertura de vértices** como un conjunto de nodos que contienen al menos un vértice incidente que conecta con cada arco de éste. Dado un grafo no dirigido $G = (V,E)$, tenemos entonces que el problema de la cobertura de vértices mínima de $G$ busca el menor subconjunto $S$ de nodos donde cada arista del grafo es incidente con al menos un nodo de $S$.

![image](https://upload.wikimedia.org/wikipedia/commons/thumb/b/bd/Minimum-vertex-cover.svg/1920px-Minimum-vertex-cover.svg.png)

Por ejemplo, en el grafo presentado abajo, los nodos ${1, 3, 5, 6}$ conforman un subconjunto que cubre cada vértice, siendo de tamaño 4. Sin embargo, este no es el menor, existiendo también las coberturas de vértices ${2,4,5}$ y ${1,2,4}$, ambas de tamaño 3.

![image](https://upload.wikimedia.org/wikipedia/commons/thumb/5/5b/6n-graf.svg/375px-6n-graf.svg.png)

###2.2.2 ¿Por qué es NP-Hard?

###Teorema (NP-Hardness de la cobertura de vértices)
> *El problema de la cobertura de vértices es NP-Hard.*

Al igual que el problema anterior, para probar este teorema aplicaremos el principio de **reducción**, donde el presente problema $B$ será el resultado de haber, tal como dice su nombre, reducido desde un problema $A$ que daremos por hecho es NP-Hard. En nuestro caso, el problema $A$ será el **problema del conjunto independiente máximo**, el cual intentaremos reducir a el problema $B$, es decir, el **problema de la cobertura de vértices**.

###Prueba del teorema
Para comenzar a comprobar si efectivamente nuestro problema $A$ se puede reducir a $B$, aunque también fue explicado con el problema anterior, debemos describir formalmente en qué consiste. Al igual que la cobertura de vértices, el problema del conjunto independiente máximo recibe un grafo no dirigido $G = (V,E)$, donde dado éste se le denomina **conjunto independiente** a aquel subconjunto de nodos $S ⊆ V$ donde no hay dos o más nodos que se conecten mediante un arco $e ⊆ E$. Por lo que el problema retornaría el mayor conjunto independiente que se puede obtener desde el grafo $G$, es decir, un valor $k$ que correspondería a la cantidad de nodos que conforman $S$.

image.png

Como podemos notar, entonces entre ambos problemas existe una similitud en sus entradas, recibiendo un grafo no dirigido, pero resolviendo tareas que podrían estar en espectros opuestos, tal como se puede ver en sus nombres (**máximo** C.I y **mínima** cobertura de vértices). 

Sin embargo, lo relevante viene si analizamos la definición de cada uno, y lo que nos ayudará luego a describir un algoritmo que pueda ser capaz de resolver el problema $A$ por medio de la solución de $B$. Y es que, si $S$ es un conjunto independiente máximo para un grafo $G = (V,E)$, entonces **no** existirá un arco $e = (u,v)$ donde ambos vértices pertenezcan a $S$. Por consiguiente, al menos uno de ellos, ya sea $u$ o $v$, pertenece en $V-S$ (que llamaremos el complemento de $S$). Así, cada arco tiene al menos un extremo en $V-S$. Por lo tanto, $V-S$ debe ser una cobertura de vértices.

Para entender mejor lo explicado recientemente, podemos tomar el siguiente grafo no dirigido de 12 nodos como ejemplo.

Si buscamos en éste su máximo conjunto independiente, veremos que $S = \{3,4,5,6,8,10,11\}$, que está conformado por un total de 7 vértices.

Ahora bien, si seguimos la lógica planteada anteriormente, entonces la mínima cobertura de vértices del grafo deberían ser los nodos restantes, es decir, $V-S = 12-7 = 5$, que estaría conformado por los vértices $\{1,2,7,9,12\}$, lo cual como podemos observar es **correcto**, puesto que cada nodo involucrado en el conjunto $V-S$ cubre la mayor cantidad de vértices posible en el grafo. Además, podemos verificar lo estipulado con respecto a los arcos. Por ejemplo, tomemos el arco $e = \{3,9\}$, y vemos como ambos no pertenecen a $S$, pero sí $9$ se encuentra en el conjunto $V-S$, mientras que $3$ al máximo conjunto independiente encontrado.

Por lo tanto, efectivamente el problema de la mínima cobertura de vértices puede ser utilizado como subrutina para el problema del máximo conjunto independiente, formando un grafo $S$ dos subconjuntos, $(V-S)$ y $S$ respectivamente. Entonces, observamos que nuestra reducción es correcta, y en consecuencia nuestro problema es NP-Hard y el teorema estipulado es **correcto**.

##2.2.3 Algoritmo sencillo para resolver el problema
Si bien nunca existirá un algoritmo que retorne una solución eficiente para el problema, podemos proponer uno que entregue un resultado aproximado. Como entrada tendría un grafo no dirigido $G = (V,E)$ y consistiría de los siguientes pasos:
1. Inicializamos un arreglo vacío $vortex\_cover$ que almacenará cada nodo que pertenezca al subconjunto que buscamos.
2. Consideramos un conjunto que contenga todos los arcos de $G$, que llamaremos $E$
3. Mientras $E$ no esté vacío, durante cada iteración elegimos un nodo arbitrario $(u,v)$ y añadimos $u$ y $v$ al arreglo $vortex\_cover$. Luego de eso, eliminamos todos los arcos que sean adyacentes a $u$ o $v$.
4. Finalmente, tan solo retornamos la longitud de $vortex\_cover$, que sería la mínima cobertura de vértices encontrada en el grafo $G$.

Para un algoritmo tan sencillo como esto, vemos que su tiempo de ejecución es de $O(V+E)$, debido a que, además de recorrer cada arco del grafo, también analizaremos una colección de sus nodos.
